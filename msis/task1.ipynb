```python
import matplotlib.pyplot as plt
```

```python
# исходные данные
x = [0, 1, 2, 3, 4, 5, 6]
n = [165, 177, 98, 49, 17, 3, 1]
```

```python
plt.title("Полигон частот")
plt.grid()
plt.plot(x, n)
plt.plot(x, n, 'o')
```

```python
plt.title("Гистограмма")
plt.grid()
plt.bar(x, n)
```

```python
w = [i / sum(n) for i in n]
print("Относительные частоты:")
[print(i) for i in w]
print("Проверка:", sum(w))
```

```python
s = 0
n_xi = []
for i in range(len(w)):
    n_xi.append(s)
    s += w[i]
n_xi.append(1)

print("Накопленные относительные частоты:")
[print(i) for i in n_xi]
print()
```

```python
plt.title("Эмпирическая функция распределения")
plt.grid()
plt.plot(x, n_xi[:-1])
```

```python
X = sum([x[i] * n[i] for i in range(len(n))]) / sum(n)
print("Выборочное среднее:")
print(X)
```

```python
D = sum([x[i] ** 2 * n[i] - X ** 2 for i in range(len(n))]) / sum(n)
print("Выборочная дисперсия:")
print(D)
```
